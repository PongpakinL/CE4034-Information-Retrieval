```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from collections import Counter
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df['count'] = df['keyword'].str.split()
```

```python
df_exploded = df.explode('count')
```

```python
x = df_exploded['count'].value_counts()
```

```python
for element, count in x.items():
    print(element)
    print(count)
```

```python
corpus = df['keyword'].to_numpy()
corpus
```

```python
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(corpus)
```

```python
vectorizer.get_feature_names_out()
```

```python
features.shape
```

```python
search_query = input('Enter your search query: ')
query_tfidf = vectorizer.transform([search_query])
```

```python
cosine_similarities = cosine_similarity(query_tfidf, features)[0]
cosine_similarities.shape
```

```python
cosine_threshold = 0
related_indices = np.where(cosine_similarities > cosine_threshold)[0]
sorted_indices = np.argsort(cosine_similarities[related_indices])[::-1]
sorted_related = related_indices[sorted_indices]
```

```python
sorted_related
```

```python
if len(sorted_indices) == 0:
    print("Sorry, no matching documents found.")
else:
    print()
    print(f"Found {len(sorted_indices)} matching documents:")
    for i in sorted_indices:
        print(df['text'][related_indices[i]])
        print(df['label'][related_indices[i]])
        print()
```

```python


# Function to count most frequent words in a column
def count_most_frequent_words(df, n=20):
    words = ' '.join(df).split()
    word_counts = Counter(words)
    most_common_words = word_counts.most_common(n)
    word_cloud_data = [{'text': word, 'weight': count} for word, count in most_common_words]
    return word_cloud_data

# Count most frequent 20 words in 'text_column' and get the data in dictionary format
word_cloud_data = count_most_frequent_words(df['keyword'][related_indices[sorted_indices]], n=20)

# Print the word cloud data
print(word_cloud_data)
```
